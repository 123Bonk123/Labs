# Data Processing
## Part one
### Laser

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit

def linear(x, k, b):
    return np.multiply(k, x) + b

df = pd.read_excel("Experimental data.xlsx",index_col=None, header=None)
tableOneLaser = np.array(df.iloc[3:11,2:4].values).T.tolist()

popt, pcov = curve_fit(linear, tableOneLaser[0], tableOneLaser[1])
plt.plot(tableOneLaser[0], linear(tableOneLaser[0], popt[0], popt[1]), 'r-', label='fit: k=%5.3f, b=%5.3f' % tuple(popt))
plt.plot(tableOneLaser[0],tableOneLaser[1],'o')
plt.legend()
plt.xlabel('P, мощность накачки')
plt.ylabel('I, ток генерации')
plt.title('Лазер')
plt.show()
```
